```sql
WITH date_summary AS (
  SELECT *,
    TO_DATE(
      CONCAT(
          CASE 
          SUBSTRING(saledate, 5, 3)
             WHEN 'Jan' THEN '01' WHEN 'Feb' THEN '02'
             WHEN 'Mar' THEN '03' WHEN 'Apr' THEN '04'
             WHEN 'May' THEN '05' WHEN 'Jun' THEN '06'
             WHEN 'Jul' THEN '07' WHEN 'Aug' THEN '08'
             WHEN 'Sep' THEN '09' WHEN 'Oct' THEN '10'
             WHEN 'Nov' THEN '11' WHEN 'Dec' THEN '12'
    END,
    SUBSTRING(saledate, 8, 19)
  ) , 'MM dd yyyy HH:mm:ss') AS date_
  FROM `workspace`.`car_sales`.`dataset1`
)
SELECT 
  DISTINCT date_,
  MONTHNAME(date_) AS month_sold,
  YEAR(date_) AS year_sold,
  sellingprice AS selling_price,
  IFNULL(make, 'Unknown') AS car_brand,
  IFNULL(model, 'Unknown') AS car_model,
  odometer AS kilometers,
  CASE
    WHEN odometer BETWEEN 0 AND 20000 THEN 'Very Low Mileage'
    WHEN odometer BETWEEN 20001 AND 40000 THEN 'Low Mileage'
    WHEN odometer BETWEEN 40001 AND 80000 THEN 'Moderate Mileage'
    WHEN odometer BETWEEN 80001 AND 120000 THEN 'Average Mileage'
    WHEN odometer BETWEEN 120001 AND 160000 THEN 'High Mileage'
    ELSE 'Very High Mileage'
  END AS mileage_group,

  IFNULL(seller, 'Unknown') AS seller,
  IFNULL(state, 'Unknown') AS location,
  COUNT(vin) AS quantity,
  IFNULL(transmission, 'Unknown') AS transmission
FROM date_summary
WHERE date_ IS NOT NULL 
  AND sellingprice IS NOT NULL
  AND odometer IS NOT NULL  
GROUP BY 1,
         3,
         4,
         sellingprice,
         make,
         model,
         odometer,
         seller,
         state,
         transmission
ORDER BY date_;
```
